```python
import pandas as pd
from sklearn.model_selection import train_test_split
import os

# Output folder
os.makedirs("CSV files", exist_ok=True)

df = pd.read_csv("plantvillage_binary_labels.csv")

# First: train + temp (val+test)
train_df, temp_df = train_test_split(
    df,
    test_size=0.3,
    stratify=df["label"],
    random_state=42
)

# Split temp into val and test (15% each of original)
val_df, test_df = train_test_split(
    temp_df,
    test_size=0.5,
    stratify=temp_df["label"],
    random_state=42
)

print("Train:", len(train_df), "Val:", len(val_df), "Test:", len(test_df))

# Save into CSV files folder
train_df.to_csv("CSV files/plantvillage_train.csv", index=False)
val_df.to_csv("CSV files/plantvillage_val.csv", index=False)
test_df.to_csv("CSV files/plantvillage_test.csv", index=False)
```
