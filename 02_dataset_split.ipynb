```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv("plantvillage_binary_labels.csv")

# First: train + temp (val+test)
train_df, temp_df = train_test_split(
    df,
    test_size=0.3,
    stratify=df["label"],
    random_state=42
)

# Then: split temp into val and test (15 percent each of original)
val_df, test_df = train_test_split(
    temp_df,
    test_size=0.5,
    stratify=temp_df["label"],
    random_state=42
)

print("Train:", len(train_df), "Val:", len(val_df), "Test:", len(test_df))

train_df.to_csv("plantvillage_train.csv", index=False)
val_df.to_csv("plantvillage_val.csv", index=False)
test_df.to_csv("plantvillage_test.csv", index=False)
```
