```python
import os
import csv

root = "datasets"

folders = [
    "BellPepper_Healthy",
    "BellPepper_BacterialSpot",
    "Tomato_Healthy",
    "Tomato_EarlyBlight",
    # "Tomato_LateBlight",
    "Potato_Healthy",
    "Potato_EarlyBlight",
    # "Potato_LateBlight",
]

rows = []

for folder in folders:
    full_folder = os.path.join(root, folder)
    crop = folder.split("_")[0].lower()          # bellpepper, tomato, potato

    if "Healthy" in folder:
        disease = "healthy"
        label = 0   # healthy
    else:
        disease = "_".join(folder.split("_")[1:]).lower()
        label = 1   # diseased

    for fname in os.listdir(full_folder):
        if not fname.lower().endswith((".jpg", ".jpeg", ".png")):
            continue

        image_path = os.path.join(full_folder, fname)
        rows.append([image_path, crop, disease, label])

# Ensure output folder exists
os.makedirs("CSV", exist_ok=True)

# Output CSV path
csv_path = os.path.join("CSV", "plantvillage_binary_labels.csv")

with open(csv_path, "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["image_path", "crop", "disease", "label"])
    writer.writerows(rows)

print("Saved:", csv_path, "with", len(rows), "rows")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Output folder
os.makedirs("CSV", exist_ok=True)

df = pd.read_csv("CSV/plantvillage_binary_labels.csv")

# First: train + temp (val+test)
train_df, temp_df = train_test_split(
    df,
    test_size=0.3,
    stratify=df["label"],
    random_state=42
)

# Split temp into val and test (15% each of original)
val_df, test_df = train_test_split(
    temp_df,
    test_size=0.5,
    stratify=temp_df["label"],
    random_state=42
)

print("Train:", len(train_df), "Val:", len(val_df), "Test:", len(test_df))

# Save into CSV folder
train_df.to_csv("CSV/plantvillage_train.csv", index=False)
val_df.to_csv("CSV/plantvillage_val.csv", index=False)
test_df.to_csv("CSV/plantvillage_test.csv", index=False)
```
